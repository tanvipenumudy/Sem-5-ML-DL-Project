```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

NN_train = [89.77, 94.60, 98.59, 88.00, 97.75, 92.38]
NN_test = [85.05, 83.45, 96.67, 82.99, 95.67, 89.87]

a = [1, 2, 3, 4, 5, 6]

plt.style.use('seaborn')
a=[str(i) for i in a]
sns.set(font_scale=1.2)
data={'Neural Network':["NN-"+str(i) for i in a]
      ,'Training':NN_train
      ,'Validation':NN_test}
df=pd.DataFrame(data)
df = pd.melt(df, id_vars="Neural Network", var_name="Training/Validation", value_name="Accuracy")
nls=sns.catplot(x='Neural Network', y='Accuracy', hue='Training/Validation', 
            data=df, kind='bar',height=6.27, aspect=13.7/6.27)
nls.set(ylim=(0,100))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

train = [98, 99.66666666666667, 96, 99.66666666666667, 100]
test = [96.75, 98, 95.77, 98, 99.66666666666667]

a = ['SVM','Logistic Regression','Decision Tree', 'KNN', 'Ensemble Model']

plt.style.use('seaborn')
sns.set(font_scale=1.2)
data={'Model':[str(i) for i in a]
      ,'Training':train
      ,'Validation':test}
df=pd.DataFrame(data)
df = pd.melt(df, id_vars="Model", var_name="Training/Validation", value_name="Accuracy")
nls=sns.catplot(x='Model', y='Accuracy', hue='Training/Validation', 
            data=df, kind='bar',height=6.27, aspect=13.7/6.27)
nls.set(ylim=(0,100))
```

```

from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
model = Sequential()
model.add(Dense(2, input_dim=1, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```
model = Sequential()
model.add(Dense(10, activation='relu', input_dim=2))
model.add(Dense(25, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(75, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(5, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
